```python
import openai
import IPython.display as display
from PIL import Image
import requests
from io import BytesIO

openai.api_key = "KEY"



def generar_idea(categoria, descripcion):
    prompt = f"""Genera una idea innovadora de startup en la industria de {categoria}.
    al momento de crearla, ten en cuenta estos aspectos especificos: {descripcion}.

    Describe el problema que resuelve, la soluci贸n que propone, su modelo de negocio y p煤blico objetivo."""

    respuesta = openai.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "system", "content": "Eres un experto en startups y emprendimiento."},
                  {"role": "user", "content": prompt}]
    )
    return respuesta.choices[0].message.content


def generar_pitch(descripcion_idea):
    prompt = f"Crea un pitch breve y persuasivo para la siguiente startup: {descripcion_idea}. " \
             "Incluye el problema, la soluci贸n, el modelo de negocio y su diferenciaci贸n."

    respuesta = openai.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "system", "content": "Eres un experto en marketing y ventas."},
                  {"role": "user", "content": prompt}]
    )
    return respuesta.choices[0].message.content


def generar_prompt_diapositiva(descripcion_idea, estilo, descripcion):
    return f"crea una disapositiva de estilo: {estilo}, sin incluir texto de ningun tipo en ella, teniendo en cuenta: {descripcion} y que sea especificamente representativo de la startup : {descripcion_idea}."

def generar_prompt_logo(descripcion_idea, estilo, descripcion):
    return f"crea un solo logo para una estartup. que sea de estilo: {estilo}, teniendo en cuenta: {descripcion} y que sea especificamente representativo de la startup : {descripcion_idea}."



def generar_imagen(prompt):
    respuesta = openai.images.generate(
        model="dall-e-3",
        prompt=prompt,
        size="1024x1024",
        n=1
    )
    return respuesta.data[0].url


categoria = input("Ingresa una categor铆a de startup: ")
descripcion = input("Agrega aqui cualquier aspecto especifico que consideres importantes para tu startup: ")
estilo = input("ingresa el estilo que te gustaria darle a tu presentacion y a tu logo: ")




idea = generar_idea(categoria, descripcion)
pitch = generar_pitch(idea)

prompt_diapositiva = generar_prompt_diapositiva(idea, estilo, descripcion)
prompt_logo = generar_prompt_logo(idea, estilo, descripcion)

imagen = generar_imagen(prompt_diapositiva)
imagendos = generar_imagen(prompt_logo)




print("\n Idea de Startup: ")
print(idea)
print("\n Pitch Generado: ")
print(pitch)

print("\n diapositiva creada para la presentacion de la startup: ")
print(imagen)
print("\n logo representativo de la startup: ")
print(imagendos)
```
